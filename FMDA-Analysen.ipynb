```python
import pandas as pd
import pandas_datareader.data as web
import numpy as np
import matplotlib.pyplot as plt
from pypfopt.efficient_frontier import EfficientFrontier
from pypfopt import risk_models
from pypfopt import expected_returns
from pypfopt import plotting

plt.style.use('seaborn')
```

```python
def read_data(data):
    df1 = []
    df = pd.DataFrame()
    if len(data) > 1:
        for f in data:
            df1.append(pd.read_csv(f + '.csv', index_col='Date'))
        df = pd.merge(df1[0], df1[1], on='Date')
        for count in range(2,len(df1)):
            df = pd.merge(df, df1[count], on='Date')
    else:
        df = pd.read_csv(data[0] + '.csv', index_col='Date')
    count = 0
    for f in data:
        df = df.rename(columns = {df.columns[count]: f})
        count += 1
    start = df.loc[df.index.str.contains('1991-01')]
    end = df.iloc[-1:]
    df = df.loc[start.index[0]:end.index[0]]
    return df


def show_graph(df, size=None, save=False, name='Graph'):
    df1 = df.pct_change() + 1
    for c in df1.columns.values:
        plt.plot(df1[c].cumprod(), label=c)
    
    plt.xticks(['1991-01', '1996-01', '2001-01', '2006-01', '2011-01', '2016-01', '2020-12'])
#     plt.xlabel('Zeit', fontsize=16)
#     plt.ylabel('Wert', fontsize=16)
#     plt.title('Wertentwicklung relativ', fontsize=20, fontweight='bold')
    plt.legend(loc='upper left', prop={'size':14})
    if save:
        plt.savefig(name, dpi=500)
    plt.show()

    
def portfolio(df, min_volatility=False, max_sharpe=False):
    global pf_return
    global pf_vola
    global pf_weights
    u = expected_returns.mean_historical_return(df, frequency=12)
    s = risk_models.sample_cov(df, frequency=12)
    ef = EfficientFrontier(u, s)
    if min_volatility:
        weights = ef.min_volatility()
    elif max_sharpe:
        weights = ef.max_sharpe()
    cleaned_weights = ef.clean_weights()
    print(cleaned_weights)
    ef.portfolio_performance(verbose=True)
    weights = list(weights.values())
    pf_weights = np.array(weights)
    pf_return = ((ef.portfolio_performance()[0]) + 1) ** (1/12) - 1
    pf_vola = ef.portfolio_performance()[1] / (12**(1/2))
    
    
def mcs(weights, port_return, port_vola, num_months=36, num_simulations=1000, start=1):
    simulation_df = []
    count = 0
    simulation_df = pd.DataFrame()
    for x in range(num_simulations):
        count = 0
        price_series = []
        price_series.append(start)
        for y in range(num_months):
            if count == num_months:
                break
            price = price_series[count] * (1 + np.random.normal(port_return, port_vola))
            price_series.append(price)
            count += 1
        simulation_df[x] = price_series  
    return simulation_df
```
